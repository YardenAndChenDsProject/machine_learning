```python
# YOUR CODE HERE
from sklearn.model_selection   import train_test_split,GridSearchCV
from sklearn.pipeline          import Pipeline
from sklearn.linear_model      import LogisticRegression
from sklearn.ensemble          import RandomForestClassifier
from sklearn.ensemble          import AdaBoostClassifier
from sklearn.tree              import DecisionTreeClassifier
from sklearn.svm               import SVC
from sklearn.metrics           import confusion_matrix,make_scorer,f1_score
from sklearn.decomposition     import TruncatedSVD
from sklearn.preprocessing     import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix
```

```python
# YOUR CODE HERE
df_train.story = df_train.story.apply(lambda x:re.sub(r'[^\w\s]','',x))
df_test.story  = df_test.story.apply(lambda x:re.sub(r'[^\w\s]','',x))
df_train.story = df_train.story.apply(lambda x:re.sub(r'\d+', '', x))
df_test.story  = df_test.story.apply(lambda x:re.sub(r'\d+', '', x))
df_train.gender = df_train.gender.replace({'m':1,'f':0})
```

```python
# Add as many code cells as you need
# YOUR CODE HERE
df_test
```

```python
df_train['num_of_Words'] = df_train.story.apply(lambda x:len(x.strip().split()))
df_test['num_of_Words']  = df_test.story.apply(lambda x:len(x.strip().split()))

df_train['avg_wrd_len'] = df_train.story.apply(lambda x:np.mean(np.array([len(word) for word in x.strip().split()])))
df_test['avg_wrd_len']  = df_test.story.apply(lambda x:np.mean(np.array([len(word) for word in x.strip().split()])))
```

```python
def count_letter_suffix(sir,letter):
    words = sir.strip().split()
    number_of_heis_at_end = 0
    for word in words:
        if word[-1]==letter:
            number_of_heis_at_end+=1
    return number_of_heis_at_end
```

```python
df_train['h_count'] = df_train.story.apply(count_letter_suffix,args=('ה'))
df_test['h_count']  = df_test.story.apply(count_letter_suffix,args=('ה'))

df_train['t_count'] = df_train.story.apply(count_letter_suffix,args=('ת'))
df_test['t_count']  = df_test.story.apply(count_letter_suffix,args=('ת'))

#Text length
df_train['txt_len'] = df_train.story.apply(lambda x: len(x))
df_test['txt_len']  = df_test.story.apply(lambda x: len(x))
```

```python
df_train
```

```python
naive = Pipeline([
    ('vect', CountVectorizer(min_df=0.05, max_df=0.7, ngram_range=(1,2), token_pattern=r'([א-ת]+)')),
    ('norm', preprocessing.Normalizer(norm='max')),
    ('clf_nb', MultinomialNB(alpha=1, fit_prior=False)),
])
rd = Pipeline([
    ('vect', CountVectorizer(min_df=0.05, max_df=0.7, ngram_range=(1,2), token_pattern=r'([א-ת]+)')),
    ('norm', preprocessing.Normalizer(norm='max')),
    ('clf_nb', RandomForestClassifier(random_state=42)),
])
svm = Pipeline([
    ('vect', CountVectorizer(min_df=0.05, max_df=0.7, ngram_range=(1,2), token_pattern=r'([א-ת]+)')),
    ('norm', preprocessing.Normalizer(norm='max')),
    ('clf_nb', AdaBoostClassifier(random_state=42)),
])
ada = Pipeline([
    ('vect', CountVectorizer(min_df=0.05, max_df=0.7, ngram_range=(1,2), token_pattern=r'([א-ת]+)')),
    ('norm', preprocessing.Normalizer(norm='max')),
    ('clf_nb', DecisionTreeClassifier(random_state=42)),
])
lg = Pipeline([
    ('vect', CountVectorizer(min_df=0.05, max_df=0.7, ngram_range=(1,2), token_pattern=r'([א-ת]+)')),
    ('norm', preprocessing.Normalizer(norm='max')),
    ('clf_nb', SVC(random_state=42)),
])


for clf,model in zip([naive,rd,svm,ada,lg],["Naive Bayes","Random Forest","SVM","AdaBoost","Logistic Regression"]):
    scores = cross_val_score(clf, X=df_train["story"], y=df_train["gender"], scoring='f1_macro', cv=5, n_jobs=6)
    print(f"The model {model} have an average score of {np.mean(scores)}.\n")
```
